# Data Validation

```python
# Check if order counts match
df["TotalChannelOrders"] = (
    df["InStoreOrders"] +
    df["UberEatsOrders"] +
    df["DoorDashOrders"] +
    df["SelfDeliveryOrders"]
)

df["OrderMismatch"] = df["MonthlyOrders"] - df["TotalChannelOrders"]

# Check share sum
df["ShareSum"] = (
    df["InStoreShare"] +
    df["UE_share"] +
    df["DD_share"] +
    df["SD_share"]
)

print(df[["OrderMismatch", "ShareSum"]].describe())
```

# Channel Volume Aggregation

```python
channel_orders = {
    "InStore": df["InStoreOrders"].sum(),
    "UberEats": df["UberEatsOrders"].sum(),
    "DoorDash": df["DoorDashOrders"].sum(),
    "SelfDelivery": df["SelfDeliveryOrders"].sum()
}

channel_df = pd.DataFrame(channel_orders.items(), columns=["Channel", "Orders"])
channel_df["MarketShare (%)"] = (channel_df["Orders"] / channel_df["Orders"].sum()) * 100

print(channel_df)
```

# Subregion Analysis

```python
subregion_channel = df.groupby("Subregion")[[
    "InStoreOrders",
    "UberEatsOrders",
    "DoorDashOrders",
    "SelfDeliveryOrders"
]].sum()

print(subregion_channel)
```

# Cuisine vs Channel Mix

```python
cuisine_channel = df.groupby("CuisineType")[[
    "InStoreOrders",
    "UberEatsOrders",
    "DoorDashOrders",
    "SelfDeliveryOrders"
]].mean()

print(cuisine_channel)
```

# Dependency Risk Analysis

```python
df["AggregatorDependence"] = df["UE_share"] + df["DD_share"]

high_risk = df[df["AggregatorDependence"] > 0.7]

print("High Risk Restaurants:", high_risk[["RestaurantName", "AggregatorDependence"]])
```

# Diversification Score

```python
df["DiversificationScore"] = 1 - (
    (df["InStoreShare"]**2 +
     df["UE_share"]**2 +
     df["DD_share"]**2 +
     df["SD_share"]**2)
)

df[["RestaurantName", "DiversificationScore"]].head()
```
